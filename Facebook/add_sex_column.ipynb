```python
detector = gender.Detector()
name_sex_dict = {}
for name in names:
    name_tokens = name.split()
    first_name = name_tokens[0]
    sex = detector.get_gender(first_name)
    if sex == "male" or sex == "mostly_male":
        name_sex_dict[name] = 0
    if sex == "female" or sex == "mostly_female":
        name_sex_dict[name] = 1
    if sex == "unknown" or sex == "andy":
        name_sex_dict[name] = -1
    
```
