```python
import pandas as pd
import csv
import copy
import altair as alt
alt.data_transformers.enable('default', max_rows=None)  #renders if the df has more than 5000 rows
pd.set_option('display.max_rows', None)
```

```python
path = "E:\Gianluca\Master Big Data Pisa\Progetto_Finale\Agricolo\Facebook\csv_docs_def\posts4.csv"
posts = pd.read_csv(path, usecols=["post_date","date_count", "likes","num_comments","num_shares"], 
                    parse_dates=["post_date"], quotechar='"', quoting=csv.QUOTE_ALL, encoding="utf-8", low_memory=False)
```

```python
def simplify_columns(row):
    if row["likes"][-1] == "K":
        row["likes"] = float(row["likes"][:-1])*1000
    else:
        row["likes"] = float(row["likes"])
    num_comments = row["num_comments"].split()[0]
    if num_comments[-1] == "K":
        row["num_comments"] = float(num_comments[:-1])*1000
    else:
        row["num_comments"] = float(num_comments)
    num_shares = row["num_shares"].split()[0]
    if num_shares[-1] == "K":
        row["num_shares"] = float(num_shares[:-1])*1000
    else:
        row["num_shares"] = float(num_shares)
    return row

# Apply the function to each row
posts2 = posts.apply(simplify_columns, axis=1)
```

```python
print(posts2)
```

```python
alt.Chart(posts2).mark_bar().encode(
    x=alt.X(field='post_date', type='temporal'),
    y=alt.Y(field='num_comments', type='quantitative')
).properties(width=1000, height=600)
```

```python
melted_posts2 = pd.melt(posts2, id_vars=["post_date", "date_count", "likes"], value_vars=['num_comments', 'num_shares'])
melted_posts2
```

```python
alt.Chart(melted_posts2).mark_bar().encode(
    #alt.X('post_date:T', bin=alt.Bin(maxbins=150)),
    alt.X('post_date:T'),
    alt.Y('value', stack='zero'),
    alt.Color('variable:N')
).properties(
    width=1000,
    height=600,
    title='Stacked Comments and Shares'
)
```
